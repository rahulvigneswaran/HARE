```python
import pandas as pd
import numpy as np

# Extracting parameters
# our_recourse_methods = ['randumb']        #['None', 'binary_search', 'binary_search_multi_iter']
# recourse_methods = ['wachter', 'growing_spheres', 'face', 'cchvae', 'clue', 'crud']                                                       #['wachter', 'growing_spheres', 'face'] #, 'actionable_recourse', 'cchvae']
# gt_perturbations = ['far']                                  #['near', 'intermediate', 'far']
# mlmodels = ['ann', 'linear']                                                        #['ann', 'linear']
# datasets = ['adult', 'compas', 'give_me_some_credit']                                             #['adult', 'compas', 'heloc']
# seeds = list(range(1, 5))

our_recourse_methods =  ['None', 'binary_search', 'binary_search_multi_iter']
recourse_methods = ['wachter', 'growing_spheres', 'face']                                                             #['wachter', 'growing_spheres', 'face'] #, 'actionable_recourse', 'cchvae']
gt_perturbations = ['far']                                  #['near', 'intermediate', 'far']
mlmodels = ['ann', 'linear']                                                         #['ann', 'linear']
datasets = ['give_me_some_credit'] #, 'give_me_some_credit']                                             #['adult', 'compas', 'heloc']
seeds = list(range(1, 5))
_K = 30
multi_iter = 5
# Initialize a dictionary to hold aggregated data
aggregated_data = {}

# Generate all combinations of parameters
for dataset in datasets:
    aggregated_data = {}
    for mlmodel in mlmodels:
        for recourse_method in recourse_methods:
            for our_recourse_method in our_recourse_methods:
                all_concat = []
                for gt_perturbation in gt_perturbations:
                    all_data = []
                    for seed in seeds:
                        file_path = f'/raid/rahulvigneswaran/tmlr/s_pear/PeAR/preferences_rebuttal_new/{recourse_method}_{our_recourse_method}_{dataset}_{mlmodel}_{gt_perturbation}_{seed}_K_{_K}_iter_{multi_iter}_prefs_1.txt'
                        with open(file_path, 'r') as file:
                            lines = file.readlines()
                            data = [line.strip().split(', ') for line in lines[1:]]
                            all_data.append(data[-1])
                    all_data = np.array(all_data).astype(float)
                    mean_data = np.mean(all_data, axis=0)
                    std_data = np.std(all_data, axis=0)
                    mean_data = mean_data[[0,1,8]]  # Reorder columns as needed
                    std_data = std_data[[0,1,8]]

                    final_data = []
                    for i in range(len(mean_data)):
                        final_data.append(mean_data[i])
                        final_data.append(std_data[i])

                    all_concat.append(final_data)
                aggregated_data[(dataset, mlmodel, recourse_method, our_recourse_method)] = np.hstack(all_concat)
    # Create a DataFrame from the aggregated data
    df = pd.DataFrame(aggregated_data.values(), index=aggregated_data.keys())

    # n = 2
    # new_index = pd.RangeIndex(len(df)*(n+1))
    # new_df = pd.DataFrame(np.nan, index=new_index, columns=df.columns)
    # ids = np.arange(len(df))*(n+1)
    # new_df.loc[ids] = df.values
    # df = new_df
    
    # Convert DataFrame to CSV
    df.to_csv(f'preferences_aggregated_data_{dataset}_new.csv', header=False)
df
```
